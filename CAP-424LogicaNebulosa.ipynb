```python
import numpy as np
A = np.zeros(10)
B = np.zeros(10)
C = np.zeros(10)

A[5] = 0.33/6
A[6] = 0.67/7
A[7] = 1/8
A[8] = 0.67/9
A[9] = 0.33/10

B[2] = 0.2/3
B[3] = 0.6/4
B[4] = 1/5
B[5] = 0.6/6
B[6] = 0.2/7

for i in range(10):
    if A[i] > B[i]:
        C[i] = A[i]
    else:
        C[i] = B[i]

print("C =")
for i in range(10):
    print(C[i], " + ")
```
